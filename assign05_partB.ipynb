````python
def compute_each_month_stats(data):
    
 #create lists to store each values we need
    highest_prices = []
    lowest_prices = []
    
    #let total_volumes and days_in_month equal 0 to add volume and count the dates each time
    total_volumes = 0
    days_in_month = 0
    
    #create days_smaller_price equal to 0 to count the number of days when open price is smalller than closing price
    days_smaller_price = 0
    
    #repeat for statement for each days by spliting the list by ',' and make lists for each day
    for days in data:
        days = days.split(',')
        
        #give each element names and set types if needed
        year = int(days[0])
        month = int(days[1])
        open_price = days[3]
        highest_price = float(days[4])
        lowest_price = float(days[5])
        closing_price = days[6]
        volume = days[8]
        
        #add values to the list I created to store values I need
        highest_prices.append(highest_price)
        lowest_prices.append(lowest_price)
        
        #count the number of days and sum the volumes to calculate total volume of the month and the average volume
        total_volumes += int(volume)
        days_in_month += 1
        average_volume = int(total_volumes/days_in_month)
        
        # count the number of days when open price is smalller than closing price by adding one each time we find such cases
        if open_price > closing_price:
            days_smaller_price += 1
    
    # sort both list to find the highest price and lowest price
    highest_prices.sort()
    lowest_prices.sort()      
    
    # return a list of year, month, highest price, lowest price, total volume, the average daily volume, 
    # number of days where the opening price is higher than the closing price
    return [(str(year), str(month), str(highest_prices[-1]), str(lowest_prices[0]), str(total_volumes), str(average_volume) , str(days_smaller_price))]

def compute_all_month_stats(filename):
    """
    function: compute_all_month_stats
    
    Given a filename
       
    Input parameter:
    ----------------
    
    * filename: Pathname for some CSV file containing stock-price
    data for twelve-month period.
    
    If the file cannot be opened, the function prints and error
    message and returns None.
    
    If the file does exist, the function assumes the data is
    organized in the comma-separated-value (CSV) format.
    
      
    Output:
    -------
    *  The only output produced is if the file cannot be opened.
    
    
    Return value:
    -------------
    A list with 12 tuples, one per month of the year, is produced.
    Each tuple consists of seven values: the year; the month;
    the highest price the stock traded in the month; the lowest price the
    stock traded in the month; the total volume of shares traded in the month;
    the average daily volume of shares traded in the month, rounded
    to the nearest integer; and the number of days where the opening price
    is higher than the closing price. Note that in this function you will
    ignore the adjusted-close values. Also: the tuples must be
    in order of data.
    """
    #If the file exists, read the file and store them into a list
    try:
        file = open(filename,'r')
        data = file.readlines()
        file.close()    
            
    #If the file cannot be opened, the function prints and error message and returns None.
    except FileNotFoundError:
        print("File Error, cannot be oppened")
        return None
    
    month_data = []
    current_month = data[1].split(',')[1]
    stats = []
    # repeat the process from the second line to the last line
    for i in range(1,len(data)):
        # for the last line, we use compute_each_month_stats(data) for the last month
        if i == len(data)-1:
            month_data.append(data[i])
            stats.append(compute_each_month_stats(month_data))
        # if we are still checking the same month's data, add the new data to month_data
        elif data[i].split(',')[1] == current_month:
            month_data.append(data[i])
        # if we are checking data of the different month from the previous data, we use compute_each_month_stats(data)
        # with what we have collected so far, and reset the month_data and add the new month's data to month_data
        else:
            stats.append(compute_each_month_stats(month_data))
            current_month = data[i].split(',')[1]
            month_data = []
            month_data.append(data[i])
    # return the list of tuples
    return(stats)



def main():
    filename = input("What is the name of the CSV file (include suffix!): ")
    all_stats = compute_all_month_stats(filename)

    if all_stats != None:
        for item in all_stats:
            for (yy, mm, hh, ll, vv, avg_vv, lower_close) in item:
                print(yy, mm, hh, ll, vv, avg_vv, lower_close, sep=",")
            

main()
````
