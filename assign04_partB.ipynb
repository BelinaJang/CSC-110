````python
def greatest_increase(pop_stats):
    """
    function: greatest_increase
    
    Given a list of tuples of the form (year, population), the
    function finds the year for which is recorded the greatest
    increase in population from the previous year.
       
    Input:
    ------
    
    * pop_stats: A list of tuples of the form (year, population).
      The tuples are sorted by year, and there are no gaps between
      years.
      
    Outputs:
    --------
    *  A pair of values consisting of the year with the greatest
       population increase, and the amount of that increase.
    """
    # set greatest_increase = o to keep the greatest increase after each calculation
    greatest_increase = 0
    # set greatest_year = 0 to keep the year when we get the greatesr increase after each calculation
    greatest_year = 0
    # set last_year_population equal to the firsr population in the list
    last_year_population = pop_stats[0][1]
    # repeat calculation for each tuple in the list 
    for (year, population) in pop_stats:
        # calculate increase and check if it's greater than the current greatest increase
        if population - last_year_population > greatest_increase:
            # if the increase is greater than the current greatest increase, let the increase be the value of the greatest increase 
            greatest_increase = population - last_year_population
            greatest_year = year
        # keep population of current tuple to use for the next calculation
        last_year_population = population
    return (greatest_year, greatest_increase)

def smallest_increase(pop_stats):
    """
    function: smallest_increase
    
    Given a list of tuples of the form (year, population), the
    function finds the year for which is recorded the smallest
    increase in population from the previous year.
       
    Input:
    ------
    
    * pop_stats: A list of tuples of the form (year, population).
      The tuples are sorted by year, and there are no gaps between
      years.
      
    Outputs:
    --------
    
    * A pair of values consisting of the year with the smallest
      population increase, and the amount of that increase.
    """
    # let smallest_increase equal to the first increase
    smallest_increase = pop_stats[0][1]-pop_stats[0][0]
    # let the smallest_year equal to the first year
    smallest_year = pop_stats[0][0]
    # let the last_year_population to the first population
    last_year_population = pop_stats[0][1]
    # repeat calculation for each tuple in the list
    for (year, population) in pop_stats:
        # For the first tuple, there's no previous term,so we skipped this tuple
        if population - last_year_population == 0:
            pass
        # calculate the increase 
        elif population - last_year_population < smallest_increase:
            # if the increase is smaller than the current smallest increase, let the increase be the value of the smallest increase
            smallest_increase = population - last_year_population
            smallest_year = year
        # keep population of current tuple to use for the next calculation
        last_year_population = population
    return (smallest_year, smallest_increase)


def average_percentage_increase(pop_stats):
    """
    function: average_percentage_increase
    
    Given a list of tuples of the form (year, population), the
    function computes the percentage change in population from
    year to year, and returns the average percentage change
    (i.e., a number from 0 to 100).
       
    Input:
    ------
    
    * pop_stats: A list of tuples of the form (year, population).
      The tuples are sorted by year, and there are no gaps between
      years.
      
    Output:
    --------
    
    * A single value that is the average of percentage changes
      from year to year (i.e., a value from 0.0 to 100.0).
    """
    # create rate_sum to add all the increase rate
    rate_sum = 0
    # count how many times you add the each increase rate
    repeat = 0
    for i in range(1,len(pop_stats)):
        # add increase rate calculated
        rate_sum += (pop_stats[i][1]-pop_stats[i-1][1])/pop_stats[i-1][1]
        # add 1 to repeat to count how many times we add increase rates
        repeat += 1
    # calculate average of increase rates
    average_increase = rate_sum/repeat
    return average_increase

def main():
    year, increase = greatest_increase(MIDYEAR_POPULATION)
    print("Greatest increase: ", year, increase)
    
    year, increase = smallest_increase(MIDYEAR_POPULATION)
    print("Smallest increase: ", year, increase)
    
    average = average_percentage_increase(MIDYEAR_POPULATION)
    print("Average percentage increase: ", format(average, ".2f"), "%", sep="")
    
    
main()
````
