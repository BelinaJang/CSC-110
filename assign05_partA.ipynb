````python
STOCKDATA_LIST_01=["2019,06,03,49.88,50.33,49.78,50.22,48.54,729600",
    "2019,06,04,50.31,50.50,50.17,50.37,48.68,888800",
    "2019,06,05,50.45,51.11,50.45,50.71,49.01,1144700",
    "2019,06,06,50.89,51.22,50.85,50.98,49.27,913200",
    "2019,06,07,50.65,50.77,49.98,50.07,48.93,881800",
    "2019,06,10,49.96,50.15,49.71,50.13,48.99,643200",
    "2019,06,11,50.21,50.26,49.92,50.02,48.88,1006000",
    "2019,06,12,50.02,50.33,49.88,50.15,49.01,696700",
    "2019,06,13,50.24,50.50,48.66,48.79,47.68,1993300",
    "2019,06,14,48.74,48.76,48.05,48.60,47.50,2125800",
    "2019,06,17,48.64,48.64,47.77,48.14,47.05,1751600",
    "2019,06,18,48.24,48.92,48.24,48.67,47.56,1879200",
    "2019,06,19,48.68,48.86,48.66,48.77,47.66,824100",
    "2019,06,20,48.77,48.77,48.18,48.50,47.40,902600",
    "2019,06,21,48.43,48.56,48.25,48.27,47.17,2825600",
    "2019,06,24,48.36,48.67,48.27,48.40,47.30,650100",
    "2019,06,25,48.29,48.54,48.15,48.31,47.21,963700",
    "2019,06,26,48.30,48.50,47.99,48.13,47.04,750600",
    "2019,06,27,48.02,48.52,47.96,48.34,47.24,1022000",
    "2019,06,28,48.30,48.61,48.15,48.41,47.31,979700"]

STOCKDATA_LIST_02=["2019,12,02,71.57,72.10,71.29,71.84,71.52,844000",
    "2019,12,03,71.74,72.16,70.89,72.05,71.73,505800",
    "2019,12,04,71.96,72.29,71.38,71.55,71.24,460200",
    "2019,12,05,71.54,72.12,71.20,71.74,71.42,409100",
    "2019,12,06,72.00,72.75,71.96,72.10,71.78,297500",
    "2019,12,09,71.89,72.23,71.77,71.96,71.64,233800",
    "2019,12,10,71.96,72.30,71.72,71.90,71.58,581200",
    "2019,12,11,71.94,72.20,71.34,71.69,71.38,518000",
    "2019,12,12,71.25,71.59,69.86,70.22,70.22,1054100",
    "2019,12,13,69.91,70.25,69.01,69.25,69.25,620200",
    "2019,12,16,69.18,69.61,69.17,69.42,69.42,292600",
    "2019,12,17,69.51,69.65,69.12,69.30,69.30,508200",
    "2019,12,18,69.32,69.32,68.31,68.78,68.78,583800",
    "2019,12,19,68.63,68.63,67.63,67.66,67.66,548600",
    "2019,12,20,67.53,68.05,67.53,67.90,67.90,1186800",
    "2019,12,23,67.68,68.04,67.50,67.68,67.68,213300",
    "2019,12,24,67.60,67.72,67.23,67.67,67.67,108600",
    "2019,12,27,67.70,67.82,67.45,67.78,67.78,144900",
    "2019,12,30,67.60,67.60,66.92,67.33,67.33,240400",
    "2019,12,31,67.06,67.24,66.62,67.00,67.00,213100"]


def compute_month_stats(data):
    """
    function: compute_month_stats
    
    Given a list of strings containing stock-trading data for one month,
    compute summary statistics for that month (year, month,
    highest price in month, lowest price in month, total volume
    for the month, average daily volume rounded to the nearest integer,
    and the number of days where the opening price greater than
    the closing price) and return all this as a tuple of the form
    (int, int, float, float, int, int, int).
       
    Input parameter:
    ----------------
    
    * data: A list of tuples where each string contains one day's
    trading information. All of the data in the list is for a
    single month. The values are separated by commas, and each day's
    list has the same number of values. The order of these values
    is: year, month, day, open price, highest price, lowest price,
    closing price, adjusted-closing price, total volume
      
    Output:
    -------
    *  There is no output to the console from this function.
    
    Return value:
    -------------
    A tuple with seven values is returned. The values are: the year;
    the month; the highest price the stock traded in the month;
    the lowest price the stock traded in the month; the total volume
    of shares traded in the month; the average daily volume of shares
    traded in the month, rounded to the nearest integer; and the number
    of days where the opening price is higher than the closing price.
    Note that in this function you will ignore the adjusted-close values.
    """
    
    #create lists to store each values we need
    highest_prices = []
    lowest_prices = []
    
    #let total_volumes and days_in_month equal 0 to add volume and count the dates each time
    total_volumes = 0
    days_in_month = 0
    
    #create days_smaller_price equal to 0 to count the number of days when open price is smalller than closing price
    days_smaller_price = 0
    
    #repeat for statement for each days by spliting the list by ',' and make lists for each day
    for days in data:
        days = days.split(',')
        
        #give each element names and set types if needed
        year = int(days[0])
        month = int(days[1])
        open_price = days[3]
        highest_price = float(days[4])
        lowest_price = float(days[5])
        closing_price = days[6]
        volume = days[8]
        
        #add values to the list I created to store values I need
        highest_prices.append(highest_price)
        lowest_prices.append(lowest_price)
        
        #count the number of days and sum the volumes to calculate total volume of the month and the average volume
        total_volumes += int(volume)
        days_in_month += 1
        average_volume = total_volumes/days_in_month
        
        # count the number of days when open price is smalller than closing price by adding one each time we find such cases
        if open_price > closing_price:
            days_smaller_price += 1
    
    # sort both list to find the highest price and lowest price
    highest_prices.sort()
    lowest_prices.sort()      
    
    # return a list of year, month, highest price, lowest price, total volume, the average daily volume, 
    # number of days where the opening price is higher than the closing price
    return [year, month, highest_prices[-1], lowest_prices[0], total_volumes, int(average_volume) , days_smaller_price]
    
def main():
    stock_data = [STOCKDATA_LIST_01, STOCKDATA_LIST_02]
    
    for month_data in stock_data:
        stats = compute_month_stats(month_data)
        for val in stats[:-1]:
            print(val, ",", sep="", end="")
        print(stats[-1])
    
main()
````
